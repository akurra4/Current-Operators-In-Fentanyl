```python
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Add edges from the DataFrame
for _, row in df.iterrows():
    email = row['Email']
    name = row['Name']
    tel = row['Tel']
    G.add_edge(email, name)
    G.add_edge(email, tel)

# Perform network analysis on the graph
# ...

# Example: Calculate the degree centrality of each node
degree_centrality = nx.degree_centrality(G)
print(degree_centrality)
```

```python
# Import the required libraries
import networkx as nx
from pyvis.network import Network

# Create a directed graph
G = nx.DiGraph()

# Add edges from the DataFrame for email and company names
for _, row in df.iterrows():
    email = row['Email']
    company = row['Name']
    G.add_edge(email, company)

# Create an interactive network visualization with cdn_resources='in_line'
nt = Network(notebook=True, cdn_resources='in_line')

# Add nodes with different colors for names and emails
for node in G.nodes:
    if node in df['Email'].values:
        nt.add_node(node, color='blue',size = 10)
    elif node in df['Name'].values:
        nt.add_node(node, color='red', size = 10)

# Add edges
for edge in G.edges:
    nt.add_edge(edge[0], edge[1])

# Set options to disable physics simulation
nt.options.physics.enabled = False

# Show the network visualization
nt.show('network_Email_Name.html')
```
