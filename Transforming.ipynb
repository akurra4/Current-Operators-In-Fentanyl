```python
import pandas as pd

cb_df = pd.read_csv('ChemBook_final.csv')
cb_df.head(5)
```

```python
#getting all the unique cas numbers grouped by name
cas = cb_df.groupby('Name')['CAS'].apply(list).reset_index(name='CAS')
```

```python
#getting all the product names grouped by cas number
names = cb_df.groupby('Name')['Product Name'].apply(list).reset_index(name='Product Name')
```

```python
#gettin all unique Tel numbers grouped by each Name
tel = cb_df.groupby('Name')['Tel'].apply(list).reset_index(name='Tel')
```

```python
#getting all unique emails grouped by Name
email = cb_df.groupby('Name')['Email'].apply(list).reset_index(name='Email')
```

```python
#getting all unique websites grouped by Name
website = cb_df.groupby('Name')['Website_URL'].apply(list).reset_index(name='Website_URL')
```

```python
#getting all unique Source_URL grouped by Name
source = cb_df.groupby('Name')['Source_URL'].apply(list).reset_index(name='Source_URL')
```

```python
#merging all the unique values into one dataframe
t3f = pd.merge(cas, names, on='Name')
t3f = pd.merge(t3f, tel, on='Name')
t3f = pd.merge(t3f, email, on='Name')
t3f = pd.merge(t3f, website, on='Name')
t3f = pd.merge(t3f, source, on='Name')
#removing_duplicates from the list of strings in each record 
t3f['CAS'] = t3f['CAS'].apply(lambda x: list(set(x)))
t3f['Product Name'] = t3f['Product Name'].apply(lambda x: list(set(x)))
t3f['Tel'] = t3f['Tel'].apply(lambda x: list(set(x)))
t3f['Email'] = t3f['Email'].apply(lambda x: list(set(x)))
t3f['Website_URL'] = t3f['Website_URL'].apply(lambda x: list(set(x)))
t3f['Source_URL'] = t3f['Source_URL'].apply(lambda x: list(set(x)))
t3f.head(5)
```

```python
t3f.to_csv('ChemBook_final_t3.csv', index=False)
```

```python
#get dictonaries of each row
t3f_dict = t3f.to_dict(orient='records')
```

```python
#merge Tel and Email into one key value pair "Contact"
for i in t3f_dict:
    i['Contact'] = {'Tel': i.pop('Tel'), 'Email': i.pop('Email')}
```

```python
cb_f = pd.DataFrame()
```

```python
cb_f.insert(0, 'CoC_ID', '')
cb_f.insert(1, 'Company_ID', '')
cb_f.insert(2, 'Attribute_ID', '')
cb_f.insert(3, 'Attribute', '')
cb_f.insert(4, 'Value', '')
```

```python
cb_name = pd.DataFrame(cb_f)
cb_product = pd.DataFrame(cb_f)
cb_source = pd.DataFrame(cb_f)
cb_web = pd.DataFrame(cb_f)
cb_con = pd.DataFrame(cb_f)
```

```python
for d in range(len(t3f_dict)):
    for k, v in t3f_dict[d].items():
        if k == 'Name':
            cb_name.at[d, 'Attribute_ID'] = 1
            cb_name.at[d, 'Attribute'] = 'Name'
            cb_name.at[d, 'Value'] = v
            cb_name.at[d, 'Company_ID'] = d+1
        elif k == 'Source_URL':
            cb_source.at[d, 'Attribute_ID'] = 26
            cb_source.at[d, 'Attribute'] = 'Information Source'
            cb_source.at[d, 'Value'] = v
            cb_source.at[d, 'Company_ID'] = d+1
        elif k == 'Product Name':
            cb_product.at[d, 'Attribute_ID'] = 3
            cb_product.at[d, 'Attribute'] = 'Products'
            cb_product.at[d, 'Value'] = v
            cb_product.at[d, 'Company_ID'] = d+1
        elif k == 'Website_URL':
            cb_web.at[d, 'Attribute_ID'] = 11
            cb_web.at[d, 'Attribute'] = 'Website'
            cb_web.at[d, 'Value'] = v
            cb_web.at[d, 'Company_ID'] = d+1
        elif k == 'Contact':
            cb_con.at[d, 'Attribute_ID'] = 10
            cb_con.at[d, 'Attribute'] = 'Contact'
            cb_con.at[d, 'Value'] = v
            cb_con.at[d, 'Company_ID'] = d+1
        else:
            continue
```

```python
# add cb_name, cb_source, cb_product, cb_web to the cb_f dataframe
cb_f = pd.concat([cb_f, cb_name, cb_product, cb_source, cb_web, cb_con])
```

```python
#sort cb_f by company id and attribute id in ascending order
cb_f = cb_f.sort_values(by=['Company_ID','Attribute_ID'], ascending=[True,True])

#reset index for cb_f
cb_f = cb_f.reset_index(drop=True)
```

```python
# make all values in value column to array of strings delimited by ; if the attribute is not a name or contact
for i in range(len(cb_f)):
    if cb_f.at[i, 'Attribute'] != 'Name' and cb_f.at[i, 'Attribute'] != 'Contact':
        cb_f.at[i, 'Value'] = ' ; '.join(str(value) for value in cb_f.at[i, 'Value'])


cb_f.to_csv('ChemBook_final_traccc.csv', index=False)
cb_f.head(25)
```
