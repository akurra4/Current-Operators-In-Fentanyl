```python
import pandas as pd
```

```python
# Read the data from the ChemBook_final.csv
cb_df = pd.read_csv('ChemBook_final.csv')
cb_df
```

```python
#Grouping all names by Website_URL
grouped = cb_df.groupby('Website_URL')['Name'].apply(list).reset_index(name='Name')
grouped
```

```python
# Extracting all data records that have the "manufacturer" in their Website_URL from grouped
manufacturer = grouped[grouped['Website_URL'].str.contains('manufacturer')]
manufacturer
```

```python
#Extracting all lists in Name column into a single list
manufacturer_list = manufacturer['Name'].tolist()
manufacturer_list

#Removing duplicates from the list
manufacturer_list = list(set([item for sublist in manufacturer_list for item in sublist]))
manufacturer_list
```

```python
#Find the Verified Date and  products sold by the Names in the manufacturer_list
verified = cb_df[cb_df['Name'].isin(manufacturer_list)]

#remove duplicates from verified and reset index
verified = verified[['Name', 'Product Name', 'CAS','Verified_On'][:]]
verified = verified.drop_duplicates()
verified = verified.reset_index(drop=True)
verified
```

```python
#Grouping all products by Name
verified_pn = verified.groupby('Name')['Product Name'].apply(list).reset_index(name='Product Name')
verified_pn['Product Name'] = verified_pn['Product Name'].astype(str)
```

```python
#grouping all CAS Numbers
verified_cas = verified.groupby('Name')['CAS'].apply(list).reset_index(name='CAS')
verified_cas['CAS'] = verified_cas['CAS'].astype(str)
```

```python
#Grouping all Verified on by Name
verified_on = verified.groupby('Name')['Verified_On'].apply(list).reset_index(name='Verified_On')

#remove duplicate values in the lists in Verified On column
verified_on['Verified_On'] = verified_on['Verified_On'].apply(lambda x: list(set(x)))
#removing all 'nan' in Verified_On lists
verified_on['Verified_On'] = verified_on['Verified_On'].apply(lambda x: [i for i in x if str(i) != 'nan'])
#Replace Empty lists with ''
verified_on['Verified_On'] = verified_on['Verified_On'].apply(lambda x: '' if x == [] else x)
```

```python
#Merging verified_pn and verified_on by Name
verified = pd.merge(verified_pn, verified_cas, on='Name')
verified_final = pd.merge(verified, verified_on, on='Name')

verified_final.to_csv('ChemBook_final_manufacturers.csv', index=False)
```
