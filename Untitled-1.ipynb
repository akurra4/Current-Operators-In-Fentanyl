```python
import pandas as pd
cb = pd.read_csv('D:/OneDrive - George Mason University - O365 Production/Semester 4/Capstone/scraping/Chemical_Book/ChemBook_final.csv')
cb_e = pd.read_csv('D:/OneDrive - George Mason University - O365 Production/Semester 4/Capstone/scraping/TraCCC\ChemBook_final_extra.csv')
```

```python
cb.tail(10)
```

```python
cb.insert(9, 'Verified_On', '')
cb.insert(11, 'Website_URL', '')
```

```python

for i in range(len(cb)):
    for j in range(len(cb_e)):
        if cb['Name'][i] == cb_e['Name'][j] and cb['CAS'][i] == cb_e['CAS'][j]:
            cb['Verified_On'][i] = cb_e['Verified_On'][j]
            cb['Website_URL'][i] = cb_e['Website_URL'][j]
```
