```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time


for i in range (len(casn)):
    for pg in range(1, 3):
        # URL of the webpage you want to scrape
        url1 = 'https://www.chemicalbook.com/ProductList_En.aspx?kwd='
        CAS = casn[i]
        url2 = '&a=China&left=True&page='
        url3 = '#J_Condition'
        url.append(url1 + CAS + url2 + str(pg) + url3)
#delete duplicate urls in the list
url = list(dict.fromkeys(url))
len(url)
```

ALL DATA EXTRACTION

```python
for u in url:
    # Initialize the Chrome driver
    service = Service(ChromeDriverManager().install())
    driver = webdriver.Chrome(service=service)
    
    # Open the webpage
    driver.get(u)
    #make the webpage to have a full screen
    driver.maximize_window()

    # Wait for the page to load completely
    time.sleep(1)
    
    # Extract the HTML content of the page
    stock_elements = driver.find_elements(By.CLASS_NAME, 'icon_prolist.stock_hw')
    empty = driver.find_elements(By.CLASS_NAME, 'SearchEmpty')
    cont = driver.find_elements(By.CLASS_NAME, 'content')
    actions = ActionChains(driver)

    # if box_ elements contains stock icon, then extract the stock information
    for s in range(len(stock_elements)):
        actions.move_to_element(stock_elements[s]).perform()
        
        # Extract the name of the company

        n = driver.find_elements(By.CLASS_NAME, 'list-title')[s].text
        names.append(n)

        # Extract the CAS number
        cn.append(u.split('kwd=')[1].split('&a')[0])

        #Take a Screenshot of the company name and stock information with a naming convention "CAS number_Company name"
        driver.save_screenshot(str(cn)+ str(n) + '.png')

        #progress bar in percentage
        print('Progress:', round((url.index(u) / len(url)) * 100, 2), '%')

        # Don't forget to close the browser!
        driver.quit()

        #set sleep time to avoid being blocked
        time.sleep(1)
```
