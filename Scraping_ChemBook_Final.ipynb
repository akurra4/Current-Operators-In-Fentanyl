```python
import pandas as pd
data = []
url = []
name = []
stock = []
ver = []
web = []
```

```python
casn = [
    "103-63-9",    # (2-Bromoethyl) Benzene (a known fentanyl precursor)
    "1609-66-1",   # Norfentanyl (NSC 89293)
    "288573-56-8", #
    "62811-39-6",   # (2-nitroprop-1-en-1-yl)benzene (1-phenyl-2-nitropropene; P2NP)
    "98-86-2"
    "244195-34-4",
    "244195-32-2",
    "90736-23-5",
    "1838-67-1",
    "1802489-71-9",
    "2309383-07-9",
    "101343-50-4",
    "120448-97-7",
    "2088842-68-4",
    "2748343-99-7",]
```

```python
#removing duplicate numbers from the list
casn = list(dict.fromkeys(casn))
print(len(casn))
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
import time


for i in range (len(casn)):
    for pg in range(1, 3):
        try:
            # URL of the webpage you want to scrape
            url1 = 'https://www.chemicalbook.com/ProductList_En.aspx?kwd='
            CAS = casn[i]
            url2 = '&a=China&left=True&page='
            url3 = '#J_Condition'
            u = url1 + CAS + url2 + str(pg) + url3
            
            # Initialize the Chrome driver
            service = Service(ChromeDriverManager().install())
            driver = webdriver.Chrome(service=service)
            
            # Open the webpage
            driver.get(u)

            # Extracting the elements information
            Box_elements = driver.find_elements(By.CLASS_NAME, 'ProLBox')

            # Initialize the ActionChains
            actions = ActionChains(driver)

            for Box in Box_elements:

                #Extracting Data from Each Box
                data.append(Box.text)
                names = Box.find_elements(By.CLASS_NAME, 'list-title')
                sec_element = Box.find_elements(By.CLASS_NAME, 'icon_prolist.SecurityPro')

                if len(names) == 0:
                    continue
                else:
                    name.append(names[0].text)

                if len(sec_element) == 0:
                    web.append('')
                    ver.append('')
                else :
                    web.append(names[0].get_attribute('href'))
                    actions.move_to_element(sec_element[0]).perform()
                    ver.append(sec_element[0].text)

                # Extract the URL
                url.append(u)

                #finding the stock element
                stock_element = Box.find_elements(By.CLASS_NAME, 'icon_prolist.stock_hw')
                
                if len(stock_element) == 0:
                    stock.append('')
                    continue

                # Hover over the stock element to display the stock information\
                actions.move_to_element(stock_element[0]).perform()

                # Extractinag  the information of the company's stock
                stock.append(stock_element[0].text)

                close_element = stock_element[0].find_elements(By.CLASS_NAME, 'close')
                try:
                    close_element[0].click()
                except:
                    print ('Missed Close')
                    pass
            # Don't forget to close the browser!
            driver.quit()

            if len(Box_elements) < 20:
                print(str(round((i/len(casn))*100, 2)) + '% completed')
                break

        except:
            print('Error at: ' + str(u))
            driver.quit()

        #progress bar as percentage of completion of the scraping
        print(str(round((i/len(casn))*100, 2)) + '% completed')

        #set sleep time to avoid being blocked
        time.sleep(0.5)
```

```python
id = []
n = []
tel = []
email = []
cas = []
pn = []
rem = []
dates = []
products = []
price = []
weights = []
verified = []
```

```python
for s in stock:
    if s != '' and 'No Data' not in s:
        dates.append(s.split('\n')[0].split('-')[1])
        products.append(s.split('\n')[1])
        price.append(s.split('\n')[3].split(' ')[1])
        weights.append(s.split('\n')[3].split(' ')[0])
    else:
        dates.append('')
        products.append('')
        price.append('')
        weights.append('')
```

```python
for v in ver:
    if v != '':
        verified.append(v.split('\n')[2].split(':')[1])
    else:
        verified.append('')
```

```python
print (len(data),len(url),len(name),len(stock),len(ver),len(web))
```

```python
for i in range(len(data)):
    id.append('ChemBook_'+str(i))
    for na in name:
        if na in data[i]:
            n.append(na)
            break

    if 'TEL:' in data[i]:
        tel.append(str(data[i].split('TEL: ')[1].split('\n')[0]))
    elif 'Tel:' in data[i]:
        tel.append(str(data[i].split('Tel: ')[1].split('\n')[0]))
    elif 'tel:' in data[i]:
        tel.append(str(data[i].split('tel: ')[1].split('\n')[0]))
    else:
        tel.append('')


    if 'Email:' in data[i]:
        email.append(str(data[i].split('Email:')[1].split('\n')[0]))
    elif 'E-mail:' in data[i]:
        email.append(str(data[i].split('E-mail:')[1].split('\n')[0]))
    elif 'e-mail:' in data[i]:
        email.append(str(data[i].split('e-mail:')[1].split('\n')[0]))
    elif 'email:' in data[i]:
        email.append(str(data[i].split('email:')[1].split('\n')[0]))
    else:
        email.append('')


    if 'CAS:' in data[i]:
        cas.append(str(data[i].split('CAS:')[1].split('\n')[0]))
    elif 'Cas:' in data[i]:
        cas.append(str(data[i].split('Cas:')[1].split('\n')[0]))
    elif 'cas:' in data[i]:
        cas.append(str(data[i].split('cas:')[1].split('\n')[0]))
    else:
        cas.append('')


    if 'Product Name:' in data[i]:
        pn.append(str(data[i].split('Product Name:')[1].split('\n')[0]))
    else:
        pn.append('')


    if 'Remarks:' in data[i]:
        rem.append(str(data[i].split('Remarks:')[1].split('\n')[0]))
    elif 'remarks:' in data[i]:
        rem.append(str(data[i].split('remarks:')[1].split('\n')[0]))
    elif 'REMARKS:' in data[i]:
        rem.append(str(data[i].split('RREMARKS:')[1].split('\n')[0]))
    else:
        rem.append('')
```

```python
print (len(id), len(n), len(tel), len(email), len(cas), len(pn), len(rem), len(dates), len(verified), len(products), len(price), len(weights))
```
